```python
#Creating a function which will plot the overall score of each player belonging to a user-input club.
def plot_overall_accto_club(club_name):
    player_names = players_analyze.loc[players_analyze['club'] == club_name]['short_name'].to_list()
    player_overall_scores = players_analyze.loc[players_analyze['club'] == club_name]['overall'].to_list()
    
    source = ColumnDataSource(data = dict(names = player_names, overall = player_overall_scores, color = ['cornflowerblue'] * len(player_overall_scores)))

    p5 = figure(x_range = player_names, height = 400, width = 800, title = "Overall Score of Players of " + club_name,
                tools = "hover", tooltips = "@names: @overall")

    p5.vbar(x = 'names', top = 'overall', width = 0.9, source = source, color = 'color')
    p5.xgrid.grid_line_color = None
    p5.y_range.start = 0

    p5.xaxis.major_label_orientation = "vertical"

    show(p5)
    
    print("The average of overall rating of players belonging to {} is {}".format(club_name, round(np.mean(players_analyze.loc[players_analyze['club'] == club_name]['overall']), 2)))
```
